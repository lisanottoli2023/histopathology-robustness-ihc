```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import os 
```

```python
path_to_train_data = "/home/lisa/Downloads/MIHIC_dataset/dataset/train"
path_to_test_data = "/home/lisa/Downloads/MIHIC_dataset/dataset/test"
path_to_validation_data = "/home/lisa/Downloads/MIHIC_dataset/dataset/val"
```

```python

def count_images_per_class(data_path):
    class_counts = {}
    for folder in os.listdir(data_path):
        folder_path = os.path.join(data_path, folder)
        if os.path.isdir(folder_path):
            class_counts[folder] = len(os.listdir(folder_path))
    return class_counts

train_counts = count_images_per_class(path_to_train_data)
test_counts = count_images_per_class(path_to_test_data)
val_counts = count_images_per_class(path_to_validation_data)


fig, axes = plt.subplots(1, 3, figsize=(18, 5))


axes[0].bar(train_counts.keys(), train_counts.values(), color='skyblue')
axes[0].set_title('Training Dataset', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Class')
axes[0].set_ylabel('Number of Images')
axes[0].tick_params(axis='x', rotation=45)

axes[1].bar(test_counts.keys(), test_counts.values(), color='lightcoral')
axes[1].set_title('Test Dataset', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Class')
axes[1].set_ylabel('Number of Images')
axes[1].tick_params(axis='x', rotation=45)

axes[2].bar(val_counts.keys(), val_counts.values(), color='lightgreen')
axes[2].set_title('Validation Dataset', fontsize=14, fontweight='bold')
axes[2].set_xlabel('Class')
axes[2].set_ylabel('Number of Images')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()


print(f"Train: {sum(train_counts.values())} total images")
print(f"Test: {sum(test_counts.values())} total images")
print(f"Validation: {sum(val_counts.values())} total images")
```
